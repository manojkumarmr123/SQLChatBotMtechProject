```python
sqlquery = """SELECT og.organization_id, count(p.product_id) as Products, count(o.order_id) as orders,count(card_id) as cards, sum(card_value) as tdv
FROM  dm_qlik_reports.organizations og 
join dm_qlik_reports.orders o on og.organization_id = o.organization_id
join dm_qlik_reports.cards c on o.order_id = c.order_id
join dm_qlik_reports.products p on o.product_id = p.product_id
group by og.organization_id"""
sqlquery
```

```python
df = pd.read_sql_query(sqlquery,cnn,index_col='organization_id')
df
```

```python
# Initialize instance of StandardScaler
t_scaler = StandardScaler()

# Fit and transform
df_scaled = t_scaler.fit_transform(df)
df_scaled
```

### KMeans Clustering

```python
t_kmeans = KMeans(n_clusters = 4, init = 'k-means++', random_state = 123)
```

```python
t_kmeans.fit(df_scaled)
df['cluster'] = t_kmeans.fit_predict(df_scaled)
```

```python
# Dictionary for labeling
label_dict = {
    3: 'High Performers',
    2: 'Moderate Performers',
    1: 'Low Performers',
    0: 'Churn Risks'
}

df['cluster_name'] = df['cluster'].map(label_dict)
```

```python
df
```

```python
# Scatterplot, colored by cluster
sns.lmplot(x='tdv', y='cards', hue='cluster_name', data=df, fit_reg=False)
```
